```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import time
```

```python
cmc = requests.get('https://coinmarketcap.com/currencies/holo/')
soup = BeautifulSoup(cmc.content, 'html.parser')
```

```python
print(soup.title)
```

```python
#use >> print(soup.prettify()) to find where price is
#print(soup.prettify())
#   <script type="application/ld+json">
   #{"@context":"http://schema.org/","@type":"ExchangeRateSpecification","url":"https://coinmarketcap.com/currencies/holo/","name":"Holo","currency":"HOT","currentExchangeRate":{"@type":"UnitPriceSpecification","price":0.02177792964907,"priceCurrency":"USD"}}
```

```python
data = soup.find('script', id="__NEXT_DATA__", type="application/json")

coins = {}

coin_data = json.loads(data.contents[0])
listings = coin_data['props']['initialState']['cryptocurrency']['listingLatest']['data']

for i in listings:
    coins[str(i('id'))] = i['slug']
```
